```python
from modules.dataloader import XView3DDN
import numpy as np
import pandas as pd
import torch
import rasterio
from rasterio.enums import Resampling
import matplotlib.pyplot as plt
import cv2
import os
from tqdm import tqdm
```

```python
for i in range(torch.cuda.device_count()):
    print(i, torch.cuda.get_device_name(i))
```

```python
device = torch.device("cuda:1")
```

```python
from torchvision.models import resnet50
from torchvision import io, transforms as T

transform = T.Normalize(mean=[0.485, 0.456, 0.406],
                                 std=[0.229, 0.224, 0.225])
```

```python
from modules.featextract import FeatureExtractor
from modules.model import RPN, DN
from modules.config import load_model_config
from modules.utils import evaluate, save_fig, convert_prob_to_image, prepare_label
from modules.losses import loss_fn_dn
```

```python
def prepare_backbone():
    backbone = resnet50(pretrained=True)
    for param in backbone.parameters():
        param.requires_grad = False
    backbone = backbone.eval()
    
    return backbone
```

```python
dn = torch.nn.Sequential(*[FeatureExtractor(load_model_config(prepare_backbone())), DN(128, 4)])
dn.load_state_dict(torch.load("/media/xview/xview3_challenge/xView3-Challenge/exps/ckpts/DN_2_trained_model_4_epochs.pth"))
dn.to(device)
for param in dn.parameters():
    param.requires_grad = False
dn.eval()
print()
```

```python
def preprocess_label(df):
    df = df.dropna(subset=["is_vessel"])
    return df

train_data_path = "/media/xview/xview3_challenge/dataset/data/train"
train_label_path = "/media/xview/xview3_challenge/dataset/labels/train.csv"

train_data = XView3DDN(model_callable=lambda x:x, background_chip_ratio=0.1, obj_size=7, threshold=0.25, overwrite=False,
                        labels_path=train_label_path, data_path=train_data_path, preprocess_label=preprocess_label)
```

```python
scene_ids = list(set([uid.split("_")[0] for uid in train_data.data_ixs]))
```

```python
def load_tiff(path, out_shape, k=(49,49)):
    data = rasterio.open(path).read(out_shape=out_shape, resampling=Resampling.bilinear).squeeze()
    
    blur_data = cv2.GaussianBlur(data, k, cv2.BORDER_DEFAULT)
    return blur_data
```

```python
def get_feat_ixs(pred, true):
    
    true_i, true_j = np.where(true != 0)
    
    pred_cls = np.argmax(pred, axis=0)
    pred_i, pred_j = np.where(pred_cls != 0)
    
    back_i, back_j = np.where(true == 0)
    back_ixs = np.random.choice(np.arange(len(back_i)), len(true_i))
    back_i, back_j = back_i[back_ixs], back_j[back_ixs]
    
    i = np.hstack([true_i[:5000], pred_i[:5000], back_i[:2000]])
    j = np.hstack([true_j[:5000], pred_j[:5000], back_j[:2000]])
    
    return i, j
```

```python
dataset = train_data
raster = "/bathymetry.tif"
data_path = "/media/xview/xview3_challenge/dataset/ensemble"

for scene_id in scene_ids:
    print(f"Processing {scene_id}:")
    uids = list(filter(lambda x: x.split("_")[0]==scene_id, train_data.data_ixs))
    scene = dataset.scenes[scene_id]
    path = scene.path+raster
#     if os.path.exists(os.path.join(data_path, scene_id)):
#                       continue
    bathymetry = load_tiff(path, out_shape=(1, scene.row, scene.col))
    
    X, y = [], []
    for uid in tqdm(uids):
        _, chip_row, chip_col = uid.split("_")
        chip_row, chip_col = int(chip_row), int(chip_col)
        
        rgb_img, flag = scene[chip_row, chip_col]

        (strt_r, strt_c), _ = scene.get_full_index(chip_row, chip_col)

        with torch.no_grad():
            rgb_img = torch.from_numpy(np.expand_dims(rgb_img, 0))

            img = transform(rgb_img)
            feat = dn[0](img.to(device))
            pred = dn[1](feat).squeeze().cpu().numpy()
            feat = feat.squeeze().cpu().numpy()
            
            labels = scene.labels[uid]
            class_labels, _ = prepare_label(labels, dataset.chipsize, 
                                            dataset.obj_size, dataset.threshold, fishing=True)
            1/0
            i_s, j_s = get_feat_ixs(pred, class_labels)
            
            try:
                f, b = feat[:, i_s, j_s], bathymetry[strt_r:strt_r+dataset.chipsize, 
                                                     strt_c:strt_c+dataset.chipsize][i_s, j_s]
            except IndexError:
                continue
            else:
                X.append(np.hstack([f.transpose((1,0)), b.reshape(-1,1)]))
                y.append(class_labels[i_s, j_s].reshape(-1,1))
    
    del bathymetry
    os.makedirs(os.path.join(data_path, scene_id), exist_ok=True)
    df = pd.DataFrame(np.vstack(X))
    df["label"] = np.vstack(y)
    df.drop_duplicates(inplace=True, ignore_index=True)
    
    df.to_csv(os.path.join(os.path.join(data_path, scene_id), "data.csv"), index=False)
```
